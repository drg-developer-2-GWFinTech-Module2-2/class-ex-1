```python
# -*- coding: utf-8 -*-
"""
Student Do: Trading Log.
This script demonstrates how to perform basic analysis of trading profits/losses
over the course of a month (20 business days).
"""

# @TODO: Initialize the metric variables
total = 0
num_days = 0
minimum = 0
maximum = 0

# @TODO: Initialize lists to hold profitable and unprofitable day profits/losses




# List of trading profits/losses
trading_pnl = [ -224,  352, 252, 354, -544,
                -650,   56, 123, -43,  254,
                 325, -123,  47, 321,  123,
                 133, -151, 613, 232, -311 ]

profitable_days = []
unprofitable_days = []

# @TODO: Iterate over each element of the list
for daily_pnl in trading_pnl:

    # @TODO: Cumulatively sum up the total and num_days
    total += daily_pnl
    num_days += 1

    # @TODO: Write logic to determine minimum and maximum values
    if daily_pnl > maximum:
        maximum = daily_pnl

    if daily_pnl < minimum:
        minimum = daily_pnl

    # @TODO: Write logic to determine profitable vs. unprofitable days
    if daily_pnl > 0:
        profitable_days.append(daily_pnl)
    if daily_pnl <= 0:
        unprofitable_days.append(daily_pnl)


# @TODO: Calculate the average
average = total / num_days

# @TODO: Calculate num_days metrics
num_profitable_days = len(profitable_days)
num_unprofitable_days = len(unprofitable_days)

# @TODO: Calculate percentage metrics
percentage_profitable_days = 100 * num_profitable_days / num_days
percentage_unprofitable_days = 100 * num_unprofitable_days / num_days

# total = sum(profitable_days) + sum(unprofitable_days)

# @TODO: Print out the summary statistics
print(f"---------Summary Statistics----------")
print(f"Number of Total Days: {num_days}")
print(f"Number of Profitable Days: {num_profitable_days}")
print(f"Number of Unprofitable Days: {num_unprofitable_days}")
print(f"Percentage of Profitable Days: {percentage_profitable_days}%")
print(f"Percentage of Unprofitable Days: {percentage_unprofitable_days}%")
print(f"-------------------------------------")
print(f"Profitable Days: {profitable_days}")
print(f"Unprofitable Days: {unprofitable_days}")
print(f"-------------------------------------")
print(f"Total Profits/Losses: {total}")
print(f"Daily Average: ${average}")
print(f"Worst Loss: ${minimum}")
print(f"Best Gain: ${maximum}")
```
